```
file_location = '/FileStore/tables/owid_covid_data.csv'

df = spark.read.csv(file_location,header=True,inferSchema=True)
```

```
display(df)
```

```
df.printSchema()
```

```
print('shape of the dataframe:',df.count(),len(df.columns))
print('number of unique countries',df.select('location').distinct().count())
print('number of unique dates',df.select('date').distinct().count())
```

```
display(df.summary())
```

```
from pyspark.sql.functions import length
globe = df.select('iso_code','date','total_cases','new_cases','new_cases_smoothed','total_deaths','new_deaths','new_deaths_smoothed','new_tests','total_tests','total_vaccinations','people_vaccinated','new_vaccinations','population')
globe = globe.filter("date == '2021-04-30'").where(length('iso_code')==3)
globe = globe.withColumn('case_to_pop',globe['total_cases']/globe['population'])
globe = globe.withColumn('death_to_pop',globe['total_deaths']/globe['population'])
globe.printSchema()
```

```
display(globe)
```

```
normalize = globe.columns[2:]
print(normalize)
```

```
for i in normalize:
  maxi = globe.groupby().max(i).first().asDict()['max('+str(i)+')']
  mini = globe.groupby().min(i).first().asDict()['min('+str(i)+')']
  globe = globe.withColumn(i,(globe[i]-mini)/(maxi-mini))
```

```
display(globe)
```

```
display(globe.orderBy(['total_cases'],ascending=False))
```

```
cont = df.select('iso_code','date','total_cases','new_cases','new_cases_smoothed','total_deaths','new_deaths','new_deaths_smoothed','total_vaccinations','people_vaccinated','new_vaccinations','population')
cont = cont.filter("date == '2021-04-30'").where(length('iso_code')!=3).filter("iso_code !='OWID_WRL'")
cont = cont.withColumn('case_to_pop',cont['total_cases']/cont['population'])
cont = cont.withColumn('death_to_pop',cont['total_deaths']/cont['population'])
display(cont)
```

```
normalize_new = cont.columns[2:]
for i in normalize_new:
  maxi = cont.groupby().max(i).first().asDict()['max('+str(i)+')']
  mini = cont.groupby().min(i).first().asDict()['min('+str(i)+')']
  cont = cont.withColumn(i,(cont[i]-mini)/(maxi-mini))
```

```
display(cont)
```

```
#effect of population density 
df_last_date = df.filter("date == '2021-04-30'").where(length('iso_code')==3)
population = df_last_date.select('total_cases','total_deaths','total_tests','total_vaccinations','stringency_index','median_age','aged_65_older','aged_70_older','gdp_per_capita','extreme_poverty','cardiovasc_death_rate','diabetes_prevalence','female_smokers','male_smokers','human_development_index','life_expectancy','handwashing_facilities','hospital_beds_per_thousand','population_density')
display(population.summary())
```

```
for i in population.columns:
  maxi = population.groupby().max(i).first().asDict()['max('+str(i)+')']
  mini = population.groupby().min(i).first().asDict()['min('+str(i)+')']
  population = population.withColumn(i,(population[i]-mini)/(maxi-mini))
```

```
display(population)
```

```
import seaborn as sn
import matplotlib.pyplot as plt
demo = population.toPandas()
```

```
fig = plt.figure(figsize=(20,20))
sn.heatmap(demo.corr(),annot=True,cmap="coolwarm")
display(fig)
```
